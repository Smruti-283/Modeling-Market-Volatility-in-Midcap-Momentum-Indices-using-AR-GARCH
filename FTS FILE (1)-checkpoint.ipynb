```python
import pandas as pd
import numpy as np



df = pd.read_excel('FINAL.xlsx', parse_dates=['Date'])

df = df.sort_values(by='Date')
# Calculate daily log returns
df['LogReturn'] = np.log(df['Close'] / df['Close'].shift(1))
df = df.dropna()

# Save cleaned data + returns to Excel
df.to_excel('nifty_midcap150_momentum50_cleaned_final.xlsx', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox

# Load the cleaned data with log returns
df = pd.read_excel('nifty_midcap150_momentum50_cleaned_final.xlsx', parse_dates=['Date'])

# Summary statistics
print("Summary Statistics of Log Returns:")
print(df['LogReturn'].describe())
print("\nSkewness:", df['LogReturn'].skew())
print("Kurtosis:", df['LogReturn'].kurtosis())

# Time series plot of log returns
plt.figure(figsize=(10, 4))
plt.plot(df['Date'], df['LogReturn'])
plt.title('Daily Log Returns')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.show()

# ACF and PACF plots
fig, ax = plt.subplots(1, 2, figsize=(15, 4))
plot_acf(df['LogReturn'], lags=40, ax=ax[0])
plot_pacf(df['LogReturn'], lags=40, ax=ax[1])
plt.show()

# Ljung-Box test for autocorrelation in returns and squared returns
lb_return = acorr_ljungbox(df['LogReturn'], lags=[10, 20, 30], return_df=True)
lb_squared = acorr_ljungbox(df['LogReturn']**2, lags=[10, 20, 30], return_df=True)

print("Ljung-Box test for log returns:")
print(lb_return)
print("\nLjung-Box test for squared log returns:")
print(lb_squared)
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
import matplotlib.pyplot as plt
# Fit AR(1) model
model = ARIMA(df['LogReturn'], order=(1,0,0))
model_fit = model.fit()
print(model_fit.summary())

# Residual diagnostics
resid = model_fit.resid

# Plot residuals ACF, PACF
fig, ax = plt.subplots(1, 2, figsize=(12,4))
plot_acf(resid, lags=40, ax=ax[0])
plot_pacf(resid, lags=40, ax=ax[1])
plt.show()

#Ljung-Box test of residuals
lb_test = acorr_ljungbox(resid, lags=[10,20,30], return_df=True)
print(lb_test)
```

```python

from arch import arch_model

# Assuming 'resid' contains the AR(1) residuals
garch = arch_model(resid, vol='Garch', p=1, q=1)
garch_fit = garch.fit()
print(garch_fit.summary())

# Extract parameters , ,  and their significance
params = garch_fit.params
print('omega:', params['omega'])
print('alpha:', params['alpha[1]'])
print('beta:', params['beta[1]'])
print('alpha + beta:', params['alpha[1]'] + params['beta[1]'])

from arch import arch_model

# Fit GARCH(1,1) to AR(1) residuals
garch_model = arch_model(resid, vol='Garch', p=1, q=1)
garch_res = garch_model.fit()
print(garch_res.summary())

# Parameters
omega = garch_res.params['omega']
alpha = garch_res.params['alpha[1]']
beta = garch_res.params['beta[1]']
print('omega:', omega)
print('alpha:', alpha)
print('beta:', beta)
print('Persistence (alpha + beta):', alpha + beta)
```

```python
# Forecast 3 days ahead from the last available date
forecast = garch_fit.forecast(horizon=3)
# Get variance forecasts (last row: forecasts for t+1, t+2, t+3)
variance_forecasts = forecast.variance.iloc[-1]
print("1-step ahead forecast (t+1):", variance_forecasts[0])
print("2-step ahead forecast (t+2):", variance_forecasts[1])
print("3-step ahead forecast (t+3):", variance_forecasts[2])

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.graphics.gofplots import qqplot

# Load cleaned data (with all needed columns)
df = pd.read_excel('nifty_midcap150_momentum50_cleaned_final.xlsx', parse_dates=['Date'])

# 1. Time Series Plot of Closing Prices
plt.figure(figsize=(10,4))
plt.plot(df['Date'], df['Close'])
plt.title('Nifty Midcap150 Momentum 50 Closing Prices')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.grid(True)
plt.show()

# 2. Time Series Plot of Log Returns
plt.figure(figsize=(10,4))
plt.plot(df['Date'], df['LogReturn'])
plt.title('Daily Log Returns')
plt.xlabel('Date')
plt.ylabel('Log Return')
plt.grid(True)
plt.show()

# 3. Histogram + KDE of Log Returns
plt.figure(figsize=(8,4))
sns.histplot(df['LogReturn'], bins=50, kde=True, color='blue')
plt.title('Histogram and KDE of Log Returns')
plt.xlabel('Log Return')
plt.ylabel('Frequency')
plt.show()

# 4. ACF and PACF plots of Log Returns
fig, ax = plt.subplots(1, 2, figsize=(15,4))
plot_acf(df['LogReturn'], lags=40, ax=ax[0])
plot_pacf(df['LogReturn'], lags=40, ax=ax[1])
plt.show()

# 5. QQ plot of Log Returns to assess normality
qqplot(df['LogReturn'], line='s')
plt.title('QQ Plot of Log Returns')
plt.show()

# Assign conditional variance from GARCH fit to dataframe
df['ConditionalVariance'] = garch_fit.conditional_volatility**2  # or .conditional_variance

# Plot fitted conditional variance
# 6. Time Series Plot of Fitted Conditional Variance (from GARCH)
plt.figure(figsize=(10,4))
plt.plot(df['Date'], df['ConditionalVariance'])
plt.title('Fitted Conditional Variance (Volatility) from GARCH(1,1)')
plt.xlabel('Date')
plt.ylabel('Conditional Variance')
plt.grid(True)
plt.show()


# 7. Plot of 1-, 2-, 3-step ahead volatility forecasts

last_date = df['Date'].iloc[-1]
forecast_var = [8.473291159858988e-05, 9.102803726499949e-05, 9.66735095459147e-05] 
forecast_dates = [last_date + pd.Timedelta(days=i) for i in range(1,4)]
plt.figure(figsize=(10,4))
plt.plot(df['Date'].tail(50), df['ConditionalVariance'].tail(50), label='Fitted Volatility')
plt.scatter(forecast_dates, forecast_var, color='red', label='Volatility Forecasts')
plt.title('Fitted and Forecasted Volatility')
plt.xlabel('Date')
plt.ylabel('Conditional Variance')
plt.legend()
plt.grid(True)
plt.show()

```
